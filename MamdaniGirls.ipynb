```python
import skfuzzy as skf
import numpy as np
import matplotlib.pyplot as plt
from membershipGroup import membershipGroup
from mamdaniSolver import MamdaniInference
import MembershipGirl
import pandas as pd

```

```python
# Age Membership
domainAge = np.arange(0.,25.1,0.1)
AgeMembership = membershipGroup()

AgeMembership.add_member('T1',domainAge,[0.,0.,1.,3.])
AgeMembership.add_member('T2',domainAge,[1.,3.,4.,6.])
AgeMembership.add_member('T3',domainAge,[4.,6.,7.,9.])
AgeMembership.add_member('T4',domainAge,[7.,9.,10.,12.])
AgeMembership.add_member('T5',domainAge,[10.,12.,15.,18.])
AgeMembership.add_member('T6',domainAge,[15.,18.,21.,24.1])

MamdaniSystem = MamdaniInference()

Umur = ['T1','T2','T3','T4','T5','T6']
Tinggi = ['SS','S','NH','H']
Berat = ['SU','U','I','PRO']
OutPut = ['SW','SW','W','W','SW','W','G','G','W','G','G','G','OB','OV','PRO','G']
counter = 0

for i in Umur:
    for j in Tinggi:
        for k in Berat:
            tmp = [i,j,k]
            MamdaniSystem.addRule(tmp,OutPut[counter])
            counter += 1
            if counter == 16:
                counter = 0
```

```python
dataRaw = pd.read_excel("CopyOfDataAsli.xlsx")
print(dataRaw)
```

```python
BoysData = dataRaw[dataRaw.Kelamin == "L"]
BoysData[BoysData.Kelamin == "L"].count()
```

```python
FuzzyInput = BoysData[["Usia","BB/U","PB/U"]]
```

```python
ZVal = []
Symbol = []


for i in FuzzyInput.index:

    def AgeCheck(Age: float):
        AgeMembership.captureInterp(Age, domainAge)
        AgeMembership.show_fuzzy_point()

        return AgeMembership.getCapturedMember()
    

    TempZVal = []
    TempSymbol = []
    MeanValue = []
    AgeMembership.captured_member = []
    AgeMembership.fuzzy_points = []
    AgeOutput = []
    AgeInput = FuzzyInput['Usia'][i]
    WeightInput = FuzzyInput['BB/U'][i]
    HeightInput = FuzzyInput['PB/U'][i]

    AgeOutput = AgeCheck(AgeInput)
    for i in range(len(AgeOutput)):
    
        if AgeOutput[i] == 'T1':
            HeightMembership = MembershipGirl.HeightMembershipT1
            WeightMembership = MembershipGirl.WeightMembershipT1
            OutputMembership = MembershipGirl.OutputMembershipT1

            DomainHeight = MembershipGirl.domainHeightT1
            DomainWeight = MembershipGirl.domainWeightT1
            DomainOutput = MembershipGirl.domainOT1

        elif AgeOutput[i] == 'T2':
            HeightMembership = MembershipGirl.HeightMembershipT2
            WeightMembership = MembershipGirl.WeightMembershipT2
            OutputMembership = MembershipGirl.OutputMembershipT2

            DomainHeight = MembershipGirl.domainHeightT2
            DomainWeight = MembershipGirl.domainWeightT2
            DomainOutput = MembershipGirl.domainOT2

        elif AgeOutput[i] == 'T3':
            HeightMembership = MembershipGirl.HeightMembershipT3
            WeightMembership = MembershipGirl.WeightMembershipT3
            OutputMembership = MembershipGirl.OutputMembershipT3

            DomainHeight = MembershipGirl.domainHeightT3
            DomainWeight = MembershipGirl.domainWeightT3
            DomainOutput = MembershipGirl.domainOT3

        elif AgeOutput[i] == 'T4':
            HeightMembership = MembershipGirl.HeightMembershipT4
            WeightMembership = MembershipGirl.WeightMembershipT4
            OutputMembership = MembershipGirl.OutputMembershipT4

            DomainHeight = MembershipGirl.domainHeightT4
            DomainWeight = MembershipGirl.domainWeightT4
            DomainOutput = MembershipGirl.domainOT4

        elif AgeOutput[i] == 'T5':
            HeightMembership = MembershipGirl.HeightMembershipT5
            WeightMembership = MembershipGirl.WeightMembershipT5
            OutputMembership = MembershipGirl.OutputMembershipT5

            DomainHeight = MembershipGirl.domainHeightT5
            DomainWeight = MembershipGirl.domainWeightT5
            DomainOutput = MembershipGirl.domainOT5

        elif AgeOutput[i] == 'T6':
            HeightMembership = MembershipGirl.HeightMembershipT6
            WeightMembership = MembershipGirl.WeightMembershipT6
            OutputMembership = MembershipGirl.OutputMembershipT6

            DomainHeight = MembershipGirl.domainHeightT6
            DomainWeight = MembershipGirl.domainWeightT6
            DomainOutput = MembershipGirl.domainOT6
        
        def fuzzyInterp(Height, Weight):
    
            HeightMembership.captureInterp(Height, DomainHeight)
            HeightMembership.show_fuzzy_point()
    
            WeightMembership.captureInterp(Weight, DomainWeight)
            WeightMembership.show_fuzzy_point()

    
        fuzzyInterp(HeightInput, WeightInput)

        MamdaniSystem.evaluateRules([AgeOutput[i]], HeightMembership.getCapturedMember(), WeightMembership.getCapturedMember())
        
        ScoreList = MamdaniSystem.getFuzzyScore([AgeMembership.fuzzy_points[i]], HeightMembership.fuzzy_points, WeightMembership.fuzzy_points, OutputMembership)
        
        ProjectedChart = MamdaniSystem.ProjectFuzzyScoreToOutput(ScoreList, OutputMembership)

        AggregatedChart = MamdaniSystem.aggregated(ProjectedChart)

        defuzzyResult = skf.defuzz(DomainOutput, AggregatedChart, 'centroid')

        OutputMembership.captureInterp(defuzzyResult, DomainOutput)
        OutInterp = OutputMembership.get_fuzzy_point()
        # print(OutInterp)
        print(defuzzyResult)
        
        TempZVal.append(defuzzyResult)

        tmpKeyOut = []
        for variable in OutInterp:
            keys = list(variable)
            tmpKeyOut.append(keys[0])
        
        stringEMPTY = ""

        if len(OutInterp) > 1:
            if OutInterp[0].get(tmpKeyOut[0]) > OutInterp[1].get(tmpKeyOut[1]):
                stringEMPTY += tmpKeyOut[0] + " | " +str(OutInterp[0].get(tmpKeyOut[0]))
                MeanValue.append(OutInterp[0].get(tmpKeyOut[0]))
        
            elif OutInterp[0].get(tmpKeyOut[0]) < OutInterp[1].get(tmpKeyOut[1]):
                stringEMPTY += tmpKeyOut[1] + " | " +str(OutInterp[1].get(tmpKeyOut[1]))
                MeanValue.append(OutInterp[1].get(tmpKeyOut[1]))
            else:
                stringEMPTY += tmpKeyOut[0] + " | " +str(OutInterp[0].get(tmpKeyOut[0])) +" || "+tmpKeyOut[1] + " | " +str(OutInterp[1].get(tmpKeyOut[1]))
                MeanValue.append(OutInterp[0].get(tmpKeyOut[0]))
        else:
            stringEMPTY += tmpKeyOut[0] + " | " + str(OutInterp[0].get(tmpKeyOut[0]))
            MeanValue.append(OutInterp[0].get(tmpKeyOut[0]))
            
        TempSymbol.append(stringEMPTY)

        OutputMembership.captured_member = []
        OutputMembership.fuzzy_points = []
        MamdaniSystem.captured_output = []
        HeightMembership.captured_member = []
        WeightMembership.captured_member= []
        HeightMembership.fuzzy_points = []
        WeightMembership.fuzzy_points= []
    
    if len(MeanValue) > 1:
        if MeanValue[0] > MeanValue[1]:
            ZVal.append(TempZVal[0])
            Symbol.append(TempSymbol[0])
        elif MeanValue[0] < MeanValue[1]:
            ZVal.append(TempZVal[1])
            Symbol.append(TempSymbol[1])
        elif MeanValue[0] == MeanValue[1]:
            ZVal.append(TempZVal[0])
            Symbol.append(TempSymbol[0])
    else:
        ZVal.append(TempZVal[0])
        Symbol.append(TempSymbol[0])
    
    CheckZval = len(ZVal)
    
    
```

```python
SlicedOutput = dataRaw[['Status gizi (metod mamdani)','Status gizi (metode tsukamoto)']]
SliceInput = dataRaw[['No','Data ke','Kelamin','Usia','BB/U','PB/U']]
SliceInput["Z Value (Mamdani)"] = np.NaN
```
