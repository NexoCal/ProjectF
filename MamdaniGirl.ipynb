```python
import skfuzzy as skf
import numpy as np
import matplotlib.pyplot as plt
from membershipGroup import membershipGroup
from mamdaniSolver import MamdaniInference
import MembershipBoy
import pandas as pd
```

```python
# Age Membership
domainAge = np.arange(0.,25.1,0.1)
AgeMembership = membershipGroup()

AgeMembership.add_member('T1',domainAge,[0.,0.,1.,3.])
AgeMembership.add_member('T2',domainAge,[1.,3.,4.,6.])
AgeMembership.add_member('T3',domainAge,[4.,6.,7.,9.])
AgeMembership.add_member('T4',domainAge,[7.,9.,10.,12.])
AgeMembership.add_member('T5',domainAge,[10.,12.,15.,18.])
AgeMembership.add_member('T6',domainAge,[15.,18.,21.,24.1])

MamdaniSystem = MamdaniInference()

Umur = ['T1','T2','T3','T4','T5','T6']
Berat = ['SU','U','I','PO']
Tinggi = ['SS','S','NH','H']
OutPut = ['SW','SW','W','W','SW','W','G','G','W','G','G','G','OB','OV','PRO','PRO']
counter = 0

for i in Umur:
    for j in Berat:
        for k in Tinggi:
            tmp = [i,j,k]
            MamdaniSystem.addRule(tmp,OutPut[counter])
            counter += 1
            if counter == 16:
                counter = 0
```

```python
dataRaw = pd.read_excel("CopyOfDataAsli.xlsx")
print(dataRaw)
```

```python
BoysData = dataRaw[dataRaw.Kelamin == "P"]
BoysData[BoysData.Kelamin == "P"].count()
```

```python
FuzzyInput = BoysData[["Usia","BB/U","PB/U"]]
```

```python
ZVal = []
Symbol = []
Stunt = []

for i in FuzzyInput.index:

    def AgeCheck(Age: float):
        AgeMembership.captureInterp(Age, domainAge)
        AgeMembership.show_fuzzy_point()

        return AgeMembership.getCapturedMember()
    

    TempZVal = []
    TempSymbol = []
    TempStunt = []
    MeanValue = []
    AgeMembership.captured_member = []
    AgeMembership.fuzzy_points = []
    AgeOutput = []
    HeightTempVal = []
    WeightTempVal = []
    AgeTempVal = []

    AgeInput = FuzzyInput['Usia'][i]
    WeightInput = FuzzyInput['BB/U'][i]
    HeightInput = FuzzyInput['PB/U'][i]

    AgeOutput = AgeCheck(AgeInput)
    for i in range(len(AgeOutput)):
    
        domainRangeMinHeight = 0.0
        domainRangeMinWeight = 0.0

        if AgeOutput[i] == 'T1':
            HeightMembership = MembershipBoy.HeightMembershipT1
            WeightMembership = MembershipBoy.WeightMembershipT1
            OutputMembership = MembershipBoy.OutputMembershipT1

            DomainHeight = MembershipBoy.domainHeightT1
            DomainWeight = MembershipBoy.domainWeightT1
            DomainOutput = MembershipBoy.domainOT1

            domainRangeMinHeight = 43.0
            domainRangeMinWeight = 1.0

        elif AgeOutput[i] == 'T2':
            HeightMembership = MembershipBoy.HeightMembershipT2
            WeightMembership = MembershipBoy.WeightMembershipT2
            OutputMembership = MembershipBoy.OutputMembershipT2

            DomainHeight = MembershipBoy.domainHeightT2
            DomainWeight = MembershipBoy.domainWeightT2
            DomainOutput = MembershipBoy.domainOT2

            domainRangeMinHeight = 54.0
            domainRangeMinWeight = 4.0

        elif AgeOutput[i] == 'T3':
            HeightMembership = MembershipBoy.HeightMembershipT3
            WeightMembership = MembershipBoy.WeightMembershipT3
            OutputMembership = MembershipBoy.OutputMembershipT3

            DomainHeight = MembershipBoy.domainHeightT3
            DomainWeight = MembershipBoy.domainWeightT3
            DomainOutput = MembershipBoy.domainOT3

            domainRangeMinHeight = 61.7
            domainRangeMinWeight = 5.0

        elif AgeOutput[i] == 'T4':
            HeightMembership = MembershipBoy.HeightMembershipT4
            WeightMembership = MembershipBoy.WeightMembershipT4
            OutputMembership = MembershipBoy.OutputMembershipT4

            DomainHeight = MembershipBoy.domainHeightT4
            DomainWeight = MembershipBoy.domainWeightT4
            DomainOutput = MembershipBoy.domainOT4

            domainRangeMinHeight = 63.4
            domainRangeMinWeight = 5.6

        elif AgeOutput[i] == 'T5':
            HeightMembership = MembershipBoy.HeightMembershipT5
            WeightMembership = MembershipBoy.WeightMembershipT5
            OutputMembership = MembershipBoy.OutputMembershipT5

            DomainHeight = MembershipBoy.domainHeightT5
            DomainWeight = MembershipBoy.domainWeightT5
            DomainOutput = MembershipBoy.domainOT5

            domainRangeMinHeight = 68.6
            domainRangeMinWeight = 6.0

        elif AgeOutput[i] == 'T6':
            HeightMembership = MembershipBoy.HeightMembershipT6
            WeightMembership = MembershipBoy.WeightMembershipT6
            OutputMembership = MembershipBoy.OutputMembershipT6

            DomainHeight = MembershipBoy.domainHeightT6
            DomainWeight = MembershipBoy.domainWeightT6
            DomainOutput = MembershipBoy.domainOT6

            domainRangeMinHeight = 74.0
            domainRangeMinWeight = 7.0

        def fuzzyInterp(Height, Weight, HeightMin, WeightMin):

            if Height < HeightMin:
                Height = HeightMin + 0.1
            if Weight < WeightMin:
                Weight = WeightMin + 0.1

            WeightMembership.captureInterp(Weight, DomainWeight)
            WeightMembership.show_fuzzy_point()

            HeightMembership.captureInterp(Height, DomainHeight)
            HeightMembership.show_fuzzy_point()

    
        fuzzyInterp(HeightInput, WeightInput, domainRangeMinHeight, domainRangeMinWeight)

        KeyHeightTemp = HeightMembership.getCapturedMember()
        KeyWeightTemp = WeightMembership.getCapturedMember()



        if len(KeyHeightTemp) > 1:
            if HeightMembership.fuzzy_points[0].get(KeyHeightTemp[0]) > HeightMembership.fuzzy_points[1].get(KeyHeightTemp[1]):
                HeightTempVal.append(HeightMembership.fuzzy_points[0].get(KeyHeightTemp[0]))
                TempStunt = KeyHeightTemp[0]
            elif HeightMembership.fuzzy_points[0].get(KeyHeightTemp[0]) < HeightMembership.fuzzy_points[1].get(KeyHeightTemp[1]):
                HeightTempVal.append(HeightMembership.fuzzy_points[1].get(KeyHeightTemp[1]))
                TempStunt = KeyHeightTemp[1]
        else:
            HeightTempVal.append(HeightMembership.fuzzy_points[0].get(KeyHeightTemp[0]))
            TempStunt = KeyHeightTemp[0]

        if len(KeyWeightTemp) > 1:
            if WeightMembership.fuzzy_points[0].get(KeyWeightTemp[0]) > WeightMembership.fuzzy_points[1].get(KeyWeightTemp[1]):
                WeightTempVal.append(WeightMembership.fuzzy_points[0].get(KeyWeightTemp[0]))
            elif WeightMembership.fuzzy_points[0].get(KeyWeightTemp[0]) < WeightMembership.fuzzy_points[1].get(KeyWeightTemp[1]):
                WeightTempVal.append(WeightMembership.fuzzy_points[1].get(KeyWeightTemp[1]))
        else:
            WeightTempVal.append(WeightMembership.fuzzy_points[0].get(KeyWeightTemp[0]))

        AgeTempVal.append(AgeMembership.fuzzy_points[i].get(AgeOutput[i]))

        MamdaniSystem.evaluateRules([AgeOutput[i]], WeightMembership.getCapturedMember(), HeightMembership.getCapturedMember())
        ScoreList = MamdaniSystem.getFuzzyScore([AgeMembership.fuzzy_points[i]], HeightMembership.fuzzy_points, WeightMembership.fuzzy_points, OutputMembership)
        ProjectedChart = MamdaniSystem.ProjectFuzzyScoreToOutput(ScoreList, OutputMembership)
        AggregatedChart = MamdaniSystem.aggregated(ProjectedChart)
        defuzzyResult = skf.defuzz(DomainOutput, AggregatedChart, 'centroid')


        OutputMembership.captureInterp(defuzzyResult, DomainOutput)
        OutInterp = OutputMembership.get_fuzzy_point()
        # print(OutInterp)
        print(defuzzyResult)
        
        TempZVal.append(defuzzyResult)

        tmpKeyOut = []
        for variable in OutInterp:
            keys = list(variable)
            tmpKeyOut.append(keys[0])
        
        stringEMPTY = ""

        if len(OutInterp) > 1:
            if OutInterp[0].get(tmpKeyOut[0]) > OutInterp[1].get(tmpKeyOut[1]):
                stringEMPTY += tmpKeyOut[0] + " | " +str(OutInterp[0].get(tmpKeyOut[0]))
                MeanValue.append(OutInterp[0].get(tmpKeyOut[0]))
        
            elif OutInterp[0].get(tmpKeyOut[0]) < OutInterp[1].get(tmpKeyOut[1]):
                stringEMPTY += tmpKeyOut[1] + " | " +str(OutInterp[1].get(tmpKeyOut[1]))
                MeanValue.append(OutInterp[1].get(tmpKeyOut[1]))
            else:
                stringEMPTY += tmpKeyOut[0] + " | " +str(OutInterp[0].get(tmpKeyOut[0])) +" || "+tmpKeyOut[1] + " | " +str(OutInterp[1].get(tmpKeyOut[1]))
                MeanValue.append(OutInterp[0].get(tmpKeyOut[0]))
        else:
            stringEMPTY += tmpKeyOut[0] + " | " + str(OutInterp[0].get(tmpKeyOut[0]))
            MeanValue.append(OutInterp[0].get(tmpKeyOut[0]))
            
        TempSymbol.append(stringEMPTY)

        OutputMembership.captured_member = []
        OutputMembership.fuzzy_points = []
        MamdaniSystem.captured_output = []
        HeightMembership.captured_member = []
        WeightMembership.captured_member= []
        HeightMembership.fuzzy_points = []
        WeightMembership.fuzzy_points= []

    Stunt.append(TempStunt[0])
    
    if len(MeanValue) > 1:
        if MeanValue[0] > MeanValue[1]:
            ZVal.append(TempZVal[0])
            Symbol.append(TempSymbol[0])
        elif MeanValue[0] < MeanValue[1]:
            ZVal.append(TempZVal[1])
            Symbol.append(TempSymbol[1])
        elif MeanValue[0] == MeanValue[1]:
            if HeightTempVal[0] > HeightTempVal[1]:
                if WeightTempVal[0] > WeightTempVal[1]:
                    ZVal.append(TempZVal[0])
                    Symbol.append(TempSymbol[0])
                elif WeightTempVal[0] < WeightTempVal[1]:
                    ZVal.append(TempZVal[0])
                    Symbol.append(TempSymbol[0])
                else:
                    ZVal.append(TempZVal[0])
                    Symbol.append(TempSymbol[0])
            
            elif HeightTempVal[0] < HeightTempVal[1]:
                if WeightTempVal[0] > WeightTempVal[1]:
                    ZVal.append(TempZVal[1])
                    Symbol.append(TempSymbol[1])
                elif WeightTempVal[0] < WeightTempVal[1]:
                    ZVal.append(TempZVal[1])
                    Symbol.append(TempSymbol[1])
                else:
                    ZVal.append(TempZVal[1])
                    Symbol.append(TempSymbol[1])
            else:
                if WeightTempVal[0] > WeightTempVal[1]:
                    ZVal.append(TempZVal[0])
                    Symbol.append(TempSymbol[0])
                elif WeightTempVal[0] < WeightTempVal[1]:
                    ZVal.append(TempZVal[1])
                    Symbol.append(TempSymbol[1])
    else:
        ZVal.append(TempZVal[0])
        Symbol.append(TempSymbol[0])
    
    CheckZval = len(ZVal)
    
    
```
